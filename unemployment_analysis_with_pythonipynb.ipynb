## ***4. Data Vizualization, Storytelling & Experimenting with charts : Understand the relationships between variables***

#### Chart - 1- Line  Chart - Estimating Employment Trends: A Time Analysis:

```python
# Grouping data by 'Date' and calculating the sum of 'Estimated_Employed' for each date
employed_over_time = df.groupby(' Date')['Estimated_Employed'].sum()

# Plotting line chart
plt.figure(figsize=(12, 6))
plt.plot(employed_over_time.index, employed_over_time.values, marker='o', linestyle='-')

# Adding title and labels
plt.title(' Employed People  Over Time')
plt.xlabel('Date')
plt.ylabel('People Employed')

# Rotating x-axis labels for better readability
plt.xticks(rotation=45)

# Displaying grid
plt.grid(True)

# Displaying the plot
plt.tight_layout()
plt.show()
```

#### Chart - 2- Histogram chart Exploring the Variation in Unemployment Rate Across India

```python
# 2. Histogram of Estimated Unemployment Rate
plt.figure(figsize=(10, 6))
sns.histplot(df['Estimated_Unemployment_Rate_(%)'], bins=20, kde=True)
plt.title('Histogram of Unemployment Rate in india')
plt.xlabel(' Unemployment Rate (%) in india')
plt.ylabel('Frequency')
plt.grid(True)
plt.show()
```

##### 1. Why did you pick the specific chart?

#### Chart 3-Pie Chart - Statewise Breakdown of Estimated Unemployment Rate by Direction

```python
# 6. Pie Chart of Estimated Unemployment Rate by States and Direction
plt.figure(figsize=(10, 10))

# Grouping data by 'States' and 'Direction' and calculating mean of 'Estimated_Unemployment_Rate_(%)'
unemployment_by_state_direction = df.groupby(['States', 'Direction'])['Estimated_Unemployment_Rate_(%)'].mean().reset_index()

# Plotting pie chart
plt.pie(unemployment_by_state_direction['Estimated_Unemployment_Rate_(%)'],
        labels=unemployment_by_state_direction.apply(lambda x: f"{x['States']} - {x['Direction']}", axis=1),
        autopct='%1.1f%%', startangle=140)

plt.title('Estimated Unemployment Rate by States and Direction')
plt.axis('equal')
plt.show()
```

Chart 4 :- Scatter Plot - Exploring the Correlation between Labour Participation and Unemployment Rates

```python
# 5. Scatter Plot of Estimated Labour Participation Rate vs. Estimated Unemployment Rate
plt.figure(figsize=(10, 6))
sns.scatterplot(x='Estimated_Labour_Participation_Rate_(%)', y='Estimated_Unemployment_Rate_(%)', data=df)
plt.title('Scatter Plot of Labour Participation Rate vs. Unemployment Rate')
plt.xlabel('Estimated Labour Participation Rate (%)')
plt.ylabel('Estimated Unemployment Rate (%)')
plt.grid(True)
plt.show()
```
